```python
# le plus grand fichier à tensor_len phrases, on cherche le multiple de "batch_size" le plus proche de ce 
# nombre par valeur supérieur : epoch_size doit etre un multiple non nul de ce nombre (pour ne pas gaspiller) 

def getEpochSize(tensor_len, batch_size):
    i = tensor_len
    while True :
        if i%batch_size == 0 :
            return i//batch_size
        i = i + 1

import io

def n_lines(file_path):
    return len(io.open(file_path, encoding='UTF-8').read().split('\n'))
```

```python
tensor_len = max(
    n_lines(file_path = "/home/jupyter/models/africa/cluster1/data/XLM_cluster1_cluster2/processed/Bafia-Bulu.Bafia.train"),
    n_lines(file_path = "/home/jupyter/models/africa/cluster1/data/XLM_cluster1_cluster2/processed/Ghomala-Limbum.Ghomala.train"),
    n_lines(file_path = "/home/jupyter/models/africa/cluster1/data/XLM_cluster1_cluster2/processed/Bulu-Limbum.Bulu.train")
)

print("tensor_len = " + str(tensor_len))
```
