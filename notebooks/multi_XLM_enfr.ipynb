```python
# le plus grand fichier à tensor_len phrases, on cherche le multiple de "batch_size" le plus proche de ce 
# nombre par valeur supérieur : epoch_size doit etre un multiple non nul de ce nombre (pour ne pas gaspiller) 

def getEpochSize(tensor_len, batch_size):
    i = tensor_len
    while True :
        if i%batch_size == 0 :
            return i//batch_size
        i = i + 1

import io

def n_lines(file_path):
    return len(io.open(file_path, encoding='UTF-8').read().split('\n'))
```

```python
tensor_len = max(
    n_lines(file_path = "/home/jupyter/models/africa/cluster1/data/multi_xlm_enfr/processed/Bafia-en.en.train"),
    n_lines(file_path = "/home/jupyter/models/africa/cluster1/data/multi_xlm_enfr/processed/Bulu-en.en.train"),
    n_lines(file_path = "/home/jupyter/models/africa/cluster1/data/multi_xlm_enfr/processed/MKPAMAN_AMVOE_Ewondo-en.en.train"),
    n_lines(file_path = "/home/jupyter/models/africa/cluster1/data/multi_xlm_enfr/processed/Bafia-fr.fr.train"),
    n_lines(file_path = "/home/jupyter/models/africa/cluster1/data/multi_xlm_enfr/processed/Bulu-fr.fr.train"),
    n_lines(file_path = "/home/jupyter/models/africa/cluster1/data/multi_xlm_enfr/processed/MKPAMAN_AMVOE_Ewondo-fr.fr.train")
)

print("tensor_len = " + str(tensor_len))
```
